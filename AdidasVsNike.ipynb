```python
#Create value for original data frame.
shoesdf = pd.read_csv('Adidas Vs Nike.csv')

print(shoesdf)
```

```python
#length of data frame
print('The data base has a shoe total of' ,len(shoesdf))
```

```python
#cleaning data
#get rid of Product ID, Description and Last Visited
best_shoesdf = shoesdf[['Product Name', 'Brand', 'Rating', 'Reviews']]
best_shoesdf
```

```python
#information about data frame
best_shoesdf.info()
```

```python
#mean of Rating and Reviews columns in original data frame
best_shoesdf.mean()
```

```python
#spliting original data frame into brand data frames to compare Nike vs Adidas.
nikeshoe = best_shoesdf.groupby(best_shoesdf.Brand)
#Using Nike as the identifier to group the data frame by.
#nike data frame is given the value nikedf
nikedf = nikeshoe.get_group("Nike")
nikedf
```

```python
#creating adidas data frame by row index due to various adidas shoe names. 
#adidas data frame is give the value adidasdf
adidasdf = best_shoesdf.iloc[:2624,:]
print("Shape of adidas dataframe - {}". format(adidasdf.shape))
```

```python
#examining head of adidas dataframe
adidasdf.head()
```

```python
#comparing length of nike and adidas dataframes
len(adidasdf)
len(nikedf)
print("The length of adidas data frame is " ,len(adidasdf))
print("The length of nike data frame is " , len(nikedf))
```

```python
#comparing avg rating of adidas shoes vs nike shoes
adidas_rating = adidasdf["Rating"].mean()
nike_rating = nikedf["Rating"].mean()
print("The average rating for adidas shoes is " ,round(adidas_rating,2))
print("The average rating for nike shoes is " ,round(nike_rating,2))
```

```python
#comparing the avg number of reviews for adidas and nike shoes
adidas_reviews = adidasdf["Reviews"].mean()
nike_reviews = nikedf["Reviews"].mean()
print("The average number of reviews for adidas shoes are ", round(adidas_reviews,2))
print("The average number of reviews for nike shoes are " , round(nike_reviews,2))
```

```python
#sorting by most reviews  and highest rated shoes for adidas data frame
most_rated = best_shoesdf.sort_values(["Rating", "Reviews"], ascending=(False, False))
print("The top 25 highest rated with most reviews shoes are " ,most_rated.head(25))
```

```python
#use head of data frame to demonstrate columns in adidas data frame
#indexing of 25 for top 25 most reviewed and highest rated.
most_rated.head(25)
```

```python
#creating descending order of the top 25 rated nike shoes in the nike data frame.
#no nikes made the list in the orignal data frame which was used in previous cell
nike_top25 = nikedf.sort_values(["Rating", "Reviews"], ascending=(False, False))
nike_top25.head(25)
```

```python
#write out csv report file comparing ratings for adidas and nike
import csv
#labeling report file
with open('ratings&reviews_report_file.csv',mode='w')as csv_file:
  #define fieldnames
  fieldnames1=['statement','output']
  writer=csv.DictWriter(csv_file, fieldnames=fieldnames1)

#create header
  writer.writeheader()
  #create rows for mean info
  writer.writerow({'statement': 'The average ratings for adidas shoes','output':'\'' + str(round(adidas_rating,2)) + '\''})
  writer.writerow({'statement': 'The average ratings for nike shoes','output': '\'' + str(round(nike_rating,2)) + '\''})
  writer.writerow({'statement': 'The average number of reviews for adidas shoes','output': '\'' + str(round(adidas_reviews,2)) + '\''})
  writer.writerow({'statement': 'The average number of reviews for nike shoes', 'output': '\'' + str(round(nike_reviews,2)) + '\''})
  writer.writerow({'statement': 'The brand with the best ratings is','output': 'Adidas'})
  writer.writerow({'statement': 'The brand that averages the most reviews is', 'output': 'Adidas'})
  writer.writerow({'statement': '168 out of the 170 best shoes were made by', 'output': 'Adidas'})
  writer.writerow({'statement': 'People really feel passionate about', 'output': 'Adidas'})
```

```python
#creating a new data frame for shoes who have High ratings and High Reviews
#Standard is set as ratings greater than 4.55 and reviews greater than 60
#data frame is given the value ninety_fourth_df for 94th percentile.
ninety_fourth_df = best_shoesdf[(best_shoesdf['Rating']>4.55) & (best_shoesdf['Reviews'] > 60.00)]
print(ninety_fourth_df)                           
```

```python
#show the nike shoes that made the list
ninety_fourth_df.tail()
```
